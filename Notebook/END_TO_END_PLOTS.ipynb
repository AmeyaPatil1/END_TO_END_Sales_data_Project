CountPlot

Countplot is used to represent the occurrence(counts) of the observation present in the categorical variable. 
It uses the concept of a bar chart for the visual depiction.

```python
#Importing Libarries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Data set
df=pd.read_csv('END_TO_END_SALES.csv')
df.head()
```

```python
#Ploting for Converted
sns.countplot(x='Converted',data=df)
plt.show()
```

```python
# As we see using value counts we will have clear idea of it
df['Converted'].value_counts()
```

In To_Be_Converted feature there is a mistake.(Coverted needs to be converted in Converted).

```python
df['To_Be_Converted'].value_counts()
```

```python
#Here we have mistake(converted=coverted)
to_be={'Coverted':'Converted'}
df['To_Be_Converted']=df['To_Be_Converted'].replace(to_be)
```

```python
#Plotting for To_Be_Converted
sns.countplot(x='To_Be_Converted',data=df)
plt.show()
```

```python
df['To_Be_Converted'].value_counts()
```

```python
#Plotting for Leadd
sns.countplot(x='Leadd',data=df)
plt.show()
```

```python
sns.countplot(x='Leadd',hue='Converted',data=df)
plt.show()
#As we can see only lead with Hot are converted(Means they had liked the product or ideas)
```
